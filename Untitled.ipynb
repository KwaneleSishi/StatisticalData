```python
import seaborn as sns
import  matplotlib.pyplot as plt
import pandas as pd 

df = pd.read_csv('C:\\Users\\HP\\Documents\\Technical Programming 2\\dataset - 2020-09-24.csv')

#Applying the default theme
sns.set_theme()

#DEMOSTRATION OF SETTING AESTHETIC STYLE
sns.set_style("whitegrid")
#Setting the context to "talk" for larger plots suitable for presentations
sns.set_context("talk")
sns.lineplot(x="Club", y="Saves", data=df)
plt.show()
```

```python
#DISTRIBUTION PLOT
sns.displot(df['Losses'])
```

```python
#DISTRIBUTION PLOT
sns.kdeplot(data=df['Losses'])
shade=True
```

```python
#DISTRIBUTION PLOT
sns.rugplot(df['Wins'])
```

```python
#CATEGORICAL PLOT
sns.boxplot(x='Position', y='Catches', data=df)
```

```python
#CATEGORICAL PLOT
sns.violinplot(x='Position', y='Shots', data=df)
```

```python
#CATEGORICAL PLOT
sns.swarmplot(x='Position', y='Tackles', data=df)
```

```python
#PAIRWISE RELATIONSHIPS
sns.pairplot(df, hue='Position')
```

```python
#PAIRWISE RELATIONSHIPS
sns.jointplot(x='Wins', y='Losses', data=df, kind='scatter')
```
